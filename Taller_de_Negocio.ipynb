Responde a las siguientes preguntas de negocio

1. ¿Cuál es la edad promedio de los clientes interesados en adquirir un certificado de depósito?

2. En una gráfica, establece la diferencia de ingresos entre clientes interesados y no interesados en el certificado de depósito

3. En un gráfico de barras, establece la distribución de estado civil entre clientes interesados en el certificado de depósito

4. ¿Cuál es el nivel educativo predominante entre clientes interesados en el certificado de depósito?

5. ¿Cuál es la relación entre estado civil vs el interés en adquirir el certificado de depósito?

6. ¿Cuál es la distribución del medio de contacto más frecuente de los clientes interesados en adquirir el certificado de depósito?

7. En un gráfico de barras horizontales responde, ¿Cuál es el tipo de ocupación de los clientes interesados en adquirir el certificado de depósito?

8. ¿Cuál es la frecuencia de contacto con los clientes interesados antes de la conversión?

9. ¿Cuál es el patrón temporal en los contactos que conducen a una mayor conversión de clientes?

```python
# importar las librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# conecto la base de datos
path = '/content/drive/MyDrive/Colab Notebooks/Data2/dataset_banco_clean.csv'
```

```python
# Se lee el archivo CVS
data = pd.read_csv(path)
```

```python
# consulta tabla
print(data.shape)
data.head()
```

```python
# se trae la descripcion de la tabla
data.describe()
```

Resouesta pregunta 1:

```python
# Edad promedio de los clientes interesados adquirir un certificado de depósito
edad_promedio_interesados = data[data['y'] == 'yes' ] ['age'].mean()
print("1. Edad promedio de los clientes interesados: ", edad_promedio_interesados)
```

Respuesta Pregunta 2:

```python
# En una gráfica, establece la diferencia de ingresos entre clientes interesados y no interesados en el certificado de depósito

sns.boxplot(x='y', y='balance', data=data)
plt.title('Diferencia de saldo promedio anual entre clientes interesados y no interesado')
plt.show()
```

Respuesta pregunta 3:

```python
# En una gráfica, establece la distribución de estado civil entre clientes interesados en el certificado de depósito


sns.countplot(x='y', hue='marital', data=data)
plt.title('Distribución de estado civil entre clientes interesados y no interesados')
plt.show()
```

Respuesta pregunta 4:

```python
# Cuál es el nivel educativo predominante entre clientes interesados en el certificado de depósito

data.education.value_counts().plot.pie()
plt.show()
```

Respuesta pregunta 5:

```python
# Cuál es la relación entre estado civil vs el interés en adquirir el certificado de depósito
sns.countplot(x='marital', hue='y', data=data)
plt.title('Relación entre estado civil vs interés en el producto')
plt.show()
```

Respuesta Pregunta 6:

```python
# Cuál es la distribución del medio de contacto más frecuente de los clientes interesados en adquirir el certificado de depósito
plt.figure(figsize=(10, 6))
sns.countplot(x='contact', hue='y', data=data)
plt.title('Distribución de medio de contacto entre clientes interesados')
plt.show()
```

Respuesta Pregunta 7:

```python
# En un gráfico de barras horizontales responde, ¿Cuál es el tipo de ocupación de los clientes
# interesados en adquirir el certificado de depósito

sns.countplot(y='job', hue='y', data=data)
plt.title('Tipo de ocupación de los clientes interesados')
plt.show()
```

Respuesta Pregunta 8:

```python
# Cuál es la frecuencia de contacto con los clientes interesados antes de la conversión
sns.countplot(x='campaign', data=data[data['y'] == 'yes'])
plt.title('Frecuencia de contacto con los clientes interesados antes de la conversión')
plt.show()
```

Respuesta Pregunta 9:

```python
# Cuál es el patrón temporal en los contactos que conducen a una mayor conversión de clientes
sns.countplot(x='month', hue='y', data=data)
plt.title('Patrón temporal en los contactos que conducen a una mayor conversión')
plt.show()
```
