```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de muestra
np.random.seed(4072023)
x = np.random.normal(size = 200)
y = np.random.normal(size = 200)
z = np.random.normal(size = 200)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

# Gráfico de dispersión 3D
ax.scatter(x, y, z)

# Equivalente a:
ax.scatter3D(x, y, z)

# plt.show()
```
